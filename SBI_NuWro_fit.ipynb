```python
import uproot
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

file_path = "sbi-results-histograms.root"
```

```python
i=7
with uproot.open(file_path) as f:
    print(f.keys())
    h_data  = f[f"data_slice_{i}"]
    h_uboone = f[f"uboone_slice_{i}"]
    h_nuwro = f[f"nuwro_slice_{i}"]
    h_sbi_data   = f[f"sbi_data_slice_{i}"]
    h_sbi_nu   = f[f"sbi_nuwro_slice_{i}"]
```

```python
sns.set_style("whitegrid")

plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["STIXGeneral"],
    "font.size": 10,
    "mathtext.fontset": "stix",
    "text.usetex": False,
})

for i in range(9):
    with uproot.open(file_path) as f:
        print(f.keys())
        h_data  = f[f"data_slice_{i}"]
        h_uboone = f[f"uboone_slice_{i}"]
        h_nuwro = f[f"nuwro_slice_{i}"]
        h_sbi_data   = f[f"sbi_data_slice_{i}"]
        h_sbi_nu   = f[f"sbi_nuwro_slice_{i}"]

    title = h_data.title
    match = re.search(r"\[(.*?),(.*?)\]", title)
    low = float(match.group(1))
    high = float(match.group(2))
    slice_label = fr"$\cos\theta \in [{low:.2f}, {high:.2f}]$"

    plt.figure(figsize=(4,3.5))

    # values and errors from ROOT histogram
    y = h_data.values()
    yerr = h_data.errors()

    # bin geometry
    edges = h_data.axes[0].edges()
    centers = 0.5 * (edges[1:] + edges[:-1])
    xerr = 0.5 * (edges[1:] - edges[:-1])

    plt.errorbar(
        centers,
        y,
        yerr=yerr,
        xerr=xerr,
        fmt='o',
        color='k',
        markersize=3,
        label="T2K data"
    )

    plt.step(centers, h_uboone.values(), where="mid",
            color='red', lw=1.8, alpha=0.7, label="MicroBooNE tune")

    plt.step(centers, h_nuwro.values(), where="mid",
            color='turquoise', lw=1.8, label="NuWro")

    plt.step(centers, h_sbi_nu.values(), where="mid",
            color='turquoise', lw=1.8, linestyle="--", label="Inferred NuWro")

    plt.step(centers, h_sbi_data.values(), where="mid",
             lw=1.8, linestyle="--", label="Inferred T2K")

    plt.xlabel(r"$p_{\mu}\,[\mathrm{GeV}]$")
    plt.ylabel(r"$\frac{d^{2}\sigma}{dp_{\mu}\, d\cos\theta_{\mu}}\;[\mathrm{cm}^{2}/(\mathrm{GeV\ nucleon})]$")
    # main legend
    main_leg = plt.legend(fontsize=8, loc="upper left")
    plt.gca().add_artist(main_leg)

    # slice label legend
    plt.legend(
    [slice_label],
#     loc="lower right",
#     loc="upper left", # slice 2 % 6
    loc="lower right", # slice 7
    frameon=False,
    handlelength=0,
    handletextpad=0)

    plt.grid(False)

    plt.tight_layout()
    plt.savefig(f"comparison_nuwro_{i}.pdf", bbox_inches="tight")
    plt.show()
```
